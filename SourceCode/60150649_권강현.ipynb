# 필요한 패키지 임포트

```python
import tensorflow as tf
from tensorflow import keras

from keras_retinanet import models
from keras_retinanet.utils.image import read_image_bgr, preprocess_image, resize_image
from keras_retinanet.utils.visualization import draw_box, draw_caption
from keras_retinanet.utils.colors import label_color

import os
import numpy as np
import pandas as pd
import csv
import cv2
import time
from PIL import Image
import glob
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

%matplotlib inline

RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
tf.random.set_seed(RANDOM_SEED)
```

```python
jpg_fileName = sorted(glob.glob("./data/train_data/*.jpg")) # 데이터 디렉토리의 이미지들이 '이름순'으로 리스트에 들어감
txt_fileName = sorted(glob.glob("./data/train_data/*.txt")) # 데이터 디렉토리의 라벨 값들이 '이름순'으로 리스트에 들어감
```

```python
dataset = dict()
dataset["image_name"] = list()
dataset["x_min"] = list()
dataset["y_min"] = list()
dataset["x_max"] = list()
dataset["y_max"] = list()
dataset["class_name"] = list()


for i in range(len(jpg_fileName)):
    dataset["image_name"].append(jpg_fileName[i])
    
    # 텍스트 파일로 저장된 라벨 값들은 "136,14,328,414/n" 이렇게 생겼기 때문에 여기서 숫자값들만 뽑아서 리스트에 담아줌
    label_box = open(txt_fileName[i], "r").read()[:-1].split(",") 
    
    x_min = int(label_box[0])
    y_min = int(label_box[1])
    width = int(label_box[2])
    height = int(label_box[3])
    
    dataset["x_min"].append(x_min)
    dataset["y_min"].append(y_min)
    dataset["x_max"].append(x_min+width)
    dataset["y_max"].append(y_min+height)
    dataset["class_name"].append("food")
```

```python
df = pd.DataFrame(dataset)
df.tail()
```

```python
def showImage(dataframeRecord):
    """우선 제대로 이미지 데이터와 라벨 데이터가 짝지어졌는지 확인하기 위해 만든 함수"""

    label_box = [dataframeRecord.x_min, dataframeRecord.y_min, dataframeRecord.x_max, dataframeRecord.y_max]

    raw_image = read_image_bgr(dataframeRecord.image_name)
    image = raw_image.copy() # 원본 이미지를 복사함
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # BGR로 로드된 이미지를 RGB로 변환힘

    draw_box(image, label_box, color=(255, 255, 0))

    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
showImage(df.iloc[5])
```

# 전처리

```python
train_df, test_df = train_test_split(df, test_size=0.2, random_state=RANDOM_SEED) # 8:2의 비율로 트레이닝셋:테스트셋 분리
```

```python
last_model = os.path.join('snapshots', sorted(os.listdir('snapshots'))[-1])
print(last_model)

model = models.load_model(last_model, backbone_name='resnet50')
model = models.convert_model(model)
```

```python
labels_to_names = pd.read_csv('classes.csv', header=None).T.loc[0].to_dict()
print(labels_to_names)
```

# 예측 능력 확인

```python
def predict(image):
    image = preprocess_image(image.copy())
    image, scale = resize_image(image)

    boxes, scores, labels = model.predict_on_batch(np.expand_dims(image, axis=0))
    boxes /= scale

    return boxes, scores, labels
```

```python
THRES_SCORE = 0.6

def makeTrainedBox(image, boxes, scores, labels):
    for box, score, label in zip(boxes[0], scores[0], labels[0]):
        if score < THRES_SCORE:
            break

        color = label_color(label)

        b = box.astype(int)
        draw_box(image, b, color=color)

        caption = "{} {:.3f}".format(labels_to_names[label], score)
        draw_caption(image, b, caption)
```

```python
def showImage2(dataframeRecord):
    """앞서 만든 showImage함수에 훈련모델이 예측한 박스를 그리는 과정만 추가한 함수"""

    label_box = [dataframeRecord.x_min, dataframeRecord.y_min, dataframeRecord.x_max, dataframeRecord.y_max]

    raw_image = read_image_bgr(dataframeRecord.image_name)
    image = raw_image.copy()
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    draw_box(image, label_box, color=(255, 255, 0)) # 라벨 정답박스를 그림(노란색 박스)

    # 추가된 부분
    boxes, scores, labels = predict(image)
    makeTrainedBox(image, boxes, scores, labels) # 훈련모델이 예측한 박스를 추가함(파란색 박스)
    
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```
